## Encoding
Converting Categorical features (e.g. ("Name" : "Yaman") to ("Name" : 1))

```python
"""
We seperate the categorical features into two parts, nominal features and Ordinal features

Nominal features have no inherent order, whilst ordinal features have inherent order

for nominal features we use onehotencoder, and for ordinal features we use labelencoder
"""

nominal_categorical = ['Vehicle Model', 'Charging Station Location']
ordinal_categorical = ["Charger Type", "User Type", "Time of Day", "Day of Week", "Long Session"]

dfenc = df.copy() #copy the dataframe to encode it

## Encode the ordinal features 
labelencoders = {}
for column in ordinal_categorical:
    le = LabelEncoder()
    dfenc[column] = le.fit_transform(dfenc[column])
    labelencoders[column] = le

## Encode the nominal features
dfenc = pd.get_dummies(dfenc, columns=nominal_categorical)

display(dfenc.head(1))
```
