```python
import joblib
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib
```

```python
splits = joblib.load("../data/preprocessed_splits.pkl")

X_train = splits["Charging Time Difference (minutes)"]["X_train"]
X_test = splits["Charging Time Difference (minutes)"]["X_test"]
y_train = splits["Charging Time Difference (minutes)"]["y_train"]
y_test = splits["Charging Time Difference (minutes)"]["y_test"]

#X_train, X_test, y_train, y_test
```

```python
model = XGBRegressor()
model.fit(X_train, y_train)
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
train_rmse = mean_squared_error(y_train, y_pred_train)
test_rmse = mean_squared_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
print("Train R2:", train_r2)
print("Test R2:", test_r2)
```

```python
joblib.dump(model, "../saved_models/time_regression.pkl")
```
