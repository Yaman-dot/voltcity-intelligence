```python
import joblib
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
splits = joblib.load("../data/preprocessed_splits.pkl")

X_train = splits["Long Session"]["X_train"]
X_test = splits["Long Session"]["X_test"]
y_train = splits["Long Session"]["y_train"]
y_test = splits["Long Session"]["y_test"]

#X_train, X_test, y_train, y_test
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)
```

```python
train_classification = classification_report(y_train, y_pred_train)
test_classification = classification_report(y_test, y_pred_test)
print("Train classification report \n", train_classification)
print("test classification report \n", test_classification)
```

```python
joblib.dump(model, "../saved_models/long_session.pkl")
```
