```python
import pandas as pd
import numpy as np
import joblib
from sklearn.ensemble import IsolationForest
from sklearn.metrics import roc_auc_score
import pickle
from sklearn.svm import OneClassSVM
from sklearn.model_selection import train_test_split
```

```python
df = joblib.load(r"..\data\preprocessed_encoded_data.pkl")
df.head()
```

```python
num_col = ["Energy Consumed (kWh)", "Charging Rate (kW)", "Day of Week", "Battery Capacity (kWh)", "State of Charge (Start %)", "Distance Driven (since last charge) (km)"]
cat_col =["User Type", "Charger Type", "Charging Station Location_Chicago", "Charging Station Location_Houston", "Charging Station Location_Los Angeles", "Charging Station Location_New York", "Charging Station Location_San Francisco"]
```

```python
X = df[num_col + cat_col]

```

```python
X_train, X_test = train_test_split(X, test_size=0.3, random_state=42)
```

```python
iso = IsolationForest(
    contamination=0.04,
    n_estimators=300,
    max_samples="auto",
    max_features=1.0,
    random_state=42
)
iso.fit(X_train)
train_scores = -iso.decision_function(X_train)
#df["score_iso"] = -iso.decision_function(X)
df
```

```python
threshold = np.percentile(train_scores, 95)
#df["anomaly_iso"] = (df["score_iso"] >= threshold).astype(int)
```

```python
X_test_synth = X_test.copy()
X_test_synth["is_synthetic"] = 0
idx = X_test_synth.sample(frac=0.03, random_state=42).index
X_test_synth.loc[idx, "is_synthetic"] = 1
X_test_synth.loc[idx, "Energy Consumed (kWh)"] += 10
X_test_synth.loc[idx, "Battery Capacity (kWh)"] += 35
test_Scores = -iso.decision_function(X_test_synth[num_col+cat_col])
```

```python
X_test_synth["score_iso"] = -iso.decision_function(X_test_synth[num_col+cat_col])

X_test_synth["anomaly_iso"] = (X_test_synth["score_iso"] >= threshold).astype(int)
```

```python
roc = roc_auc_score(X_test_synth["is_synthetic"], X_test_synth["score_iso"])
print("ROC AUC:", roc)
```

## SVM

```python
ocsvm = OneClassSVM(
    kernel="rbf",
    nu=0.04,
    gamma=0.01,
)
ocsvm.fit(X_train)
```

```python
X_test_synth["score_ocsvm"] = -ocsvm.decision_function(X_test_synth[num_col + cat_col])
X_test_synth["anomaly_ocsvm"] = (ocsvm.predict(X_test_synth[num_col+cat_col]) == -1).astype(int)
X_test_synth.head()
```

```python
for g in ["scale", 0.01, 0.1]:
    oc = OneClassSVM(nu=0.04, gamma=g)
    oc.fit(X_train)
    scores = -oc.decision_function(X_test_synth[num_col+cat_col])
    print(f"gamma: {g}, roc: {roc_auc_score(X_test_synth["is_synthetic"], scores)}")
```
